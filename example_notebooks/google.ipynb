```python
import os
```

```python
def get_url(looking_for):
    url ="https://www.googleapis.com/customsearch/v1?" + "&".join([
        f"key={os.environ["GOOGLE_API_KEY"]}",
        f"cx={os.environ["GOOGLE_PSE_CX"]}",
        f"q={looking_for.replace(" ", "+", -1)}+filetype%3Apdf"
    ])
    return url

url = get_url("US BANK 2024 Q1 Finanical statements")
```

```python
import requests
import json
```

```python
resp = requests.get(url)
```

```python
json =json.loads(resp.content.decode("utf-8"))
```
