```python
abnb = yf.Ticker("ABNB")

financials_all = set(((lambda df: df[~(df.isna().any(axis=1))])(abnb.financials)).index)
income_statement_all = set(((lambda df: df[~(df.isna().any(axis=1))])(abnb.income_stmt)).index)
balance_sheet_all = set(((lambda df: df[~(df.isna().any(axis=1))])(abnb.balance_sheet.drop(["2019-12-31"], axis=1))).index) 
cashflow_all = set(((lambda df: df[~(df.isna().any(axis=1))])(abnb.balance_sheet.drop(["2019-12-31"], axis=1))).index)
financials_any = set(abnb.financials.index)
income_statement_any = set(abnb.income_stmt.index)
balance_sheet_any = set(abnb.balance_sheet.index)
cashflow_any = set(abnb.cashflow.index)

if(financials_all and income_statement_all and balance_sheet_all and cashflow_all and financials_any and income_statement_any and balance_sheet_any and cashflow_any):
    print('okay')
```

```python
def filter_na_row(df) -> set(str):
    return set(df[~(df.isna().any(axis=1))].index)
```

```python

for ticker in GICS.values():
    t = yf.Ticker(ticker)
    financials_any |= set(t.financials.index)
    income_statement_any |= set(t.income_stmt.index)
    balance_sheet_any |= set(t.balance_sheet.index)
    cashflow_any |= set(t.cashflow.index)

    if(f:=filter_na_row(t.financials)):
        financials_all &= f
    if(i:=filter_na_row(t.income_stmt)):
        income_statement_all &= i
    if(b:=filter_na_row(t.balance_sheet)):
        balance_sheet_all &= b
    if(c:=filter_na_row(t.cashflow)):
        cashflow_all &= c
```
